```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
file_path = "personal_cleaned_final.csv"  # Replace with your actual file path
df = pd.read_csv(file_path)

# Remove "PERSONAL_" prefix from all column names
df.columns = [col.replace("PERSONAL_", "") for col in df.columns]

# Replace "Chiron" with "SBC" in the "controller" column
df['controller'] = df['controller'].replace("Chiron", "SBC")

# Save the cleaned file
cleaned_file_path = "personal_cleaned.csv"  # Replace with desired output path
df.to_csv(cleaned_file_path, index=False)

print("Column names cleaned and file saved as", cleaned_file_path)

# Define colors for controllers
controller_colors = {"SBC": "red", "WBC": "lightgreen"}

# Define the columns to plot
questions = {
    "Is the person testing first the VR or the screen?": "Is the person testing first the VR or the screen?",
    "What is the highest degree or level of education you have completed or are currently pursuing?": "What is the highest degree or level of education you have completed or are currently pursuing?",
    "Are you wearing glasses or lenses or nothing today?": "Are you wearing glasses or lenses or nothing today?",
    "How often have you used used virtual reality gear?": "How often have you used used virtual reality gear?",
    "How often have you interacted with teleoperation setups until now?": "How often have you interacted with teleoperation setups until now?",
}

# Define specific order for frequency-related questions
frequency_order = ["Never", "1 or 2 times", "More than 2 times, but not often"]

# Create a figure for barplots with 2 plots per row
fig, axes = plt.subplots((len(questions) + 1) // 2, 2, figsize=(15, 13))
axes = axes.flatten()

# Create barplots for each question, grouped by controller
for i, (col, title) in enumerate(questions.items()):
    order = frequency_order if "virtual reality gear" in col or "teleoperation setups" in col else None
    sns.countplot(x=col, hue="controller", data=df, ax=axes[i], palette=controller_colors, order=order)
    axes[i].set_title(title)
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Count")
    axes[i].set_ylim(0, 9)  # Set y-axis limit to 0-8
    axes[i].tick_params(axis='x', rotation=0)

# Remove extra subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout and show the plots
plt.tight_layout()
plt.tight_layout()
plt.savefig("personal_summary.pdf", format='pdf')
plt.show()

# Create pie charts for field of study/work separated by controller
fig, axes = plt.subplots(2, 1, figsize=(10, 9))

for i, controller in enumerate(["WBC", "SBC"]):
    field_counts = df[df["controller"] == controller]["What is your field of study/work?"].value_counts()
    axes[i].pie(field_counts, labels=field_counts.index, autopct='%1.1f%%', startangle=140)
    axes[i].set_title(f"Field of Study/Work Distribution - {controller}")
    axes[i].legend(title="Fields of Study/Work", loc="best")
    axes[i].axis("equal")  # Equal aspect ratio ensures that pie is drawn as a circle

plt.tight_layout()
plt.tight_layout()
plt.savefig("field_work.pdf", format='pdf')
plt.show()


# Create a separate horizontal bar plot for video game experience in order
experience_order = [
    "Advance experience, played/play very often",
    "Moderate experience, played/play quite often",
    "A little experience, played a few times",
        "I have no experience, never played"
]

plt.figure(figsize=(8, 4))
sns.countplot(y="How experient do you consider your self in videogames? (Consider how often you play/have played)", hue="controller", data=df, palette=controller_colors, order=experience_order)
plt.title("How experient do you consider your self in videogames?")
plt.xlabel("Count")
plt.ylabel("")
plt.xlim(0, 8)  # Set x-axis limit to 0-8
plt.legend(title="Controller")
plt.tight_layout()
plt.tight_layout()
plt.savefig("videogame.pdf", format='pdf')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
file_path = "personal_cleaned_final.csv"  # Replace with your actual file path
df = pd.read_csv(file_path)

# Remove "PERSONAL_" prefix from all column names
df.columns = [col.replace("PERSONAL_", "") for col in df.columns]

# Replace "Chiron" with "SBC" in the "controller" column
df['controller'] = df['controller'].replace("Chiron", "SBC")

# Save the cleaned file
cleaned_file_path = "personal_cleaned.csv"  # Replace with desired output path
df.to_csv(cleaned_file_path, index=False)

print("Column names cleaned and file saved as", cleaned_file_path)

# Define colors for controllers
controller_colors = {"SBC": "red", "WBC": "lightgreen"}

# Define the columns to plot
questions = {
    "Is the person testing first the VR or the screen?": "Is the person testing first the VR or the screen?",
    "What is the highest degree or level of education you have completed or are currently pursuing?": "What is the highest degree or level of education you have completed or are currently pursuing?",
    "Are you wearing glasses or lenses or nothing today?": "Are you wearing glasses or lenses or nothing today?",
    "How often have you used used virtual reality gear?": "How often have you used used virtual reality gear?",
    "How often have you interacted with teleoperation setups until now?": "How often have you interacted with teleoperation setups until now?",
}

# Define specific order for frequency-related questions
frequency_order = ["Never", "1 or 2 times", "More than 2 times, but not often"]

# Create a figure for barplots with 2 plots per row
fig, axes = plt.subplots((len(questions) + 1) // 2, 2, figsize=(14, 18))
axes = axes.flatten()

# Create barplots for each question, grouped by controller
for i, (col, title) in enumerate(questions.items()):
    order = frequency_order if "virtual reality gear" in col or "teleoperation setups" in col else None
    sns.countplot(x=col, hue="controller", data=df, ax=axes[i], palette=controller_colors, order=order)
    axes[i].set_title(title, pad=20)
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Count")
    axes[i].set_ylim(0, 8)  # Set y-axis limit to 0-8
    axes[i].tick_params(axis='x', rotation=45)

# Remove extra subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout and show the plots
plt.tight_layout()
plt.show()

# Create pie charts for field of study/work separated by controller
fig, axes = plt.subplots(2, 1, figsize=(10, 16))

for i, controller in enumerate(["WBC", "SBC"]):
    field_counts = df[df["controller"] == controller]["What is your field of study/work?"].value_counts()
    axes[i].pie(field_counts, labels=field_counts.index, autopct='%1.1f%%', startangle=140)
    axes[i].set_title(f"Field of Study/Work Distribution - {controller}", pad=20)
    axes[i].axis("equal")  # Equal aspect ratio ensures that pie is drawn as a circle

plt.tight_layout()
plt.show()

# Create a separate horizontal bar plot for video game experience in order
experience_order = [
    "I have no experience, never played",
    "I have played a few times",
    "I play occasionally",
    "I play frequently",
    "I am an advanced player"
]

plt.figure(figsize=(10, 6))
sns.countplot(y="How experient do you consider your self in videogames? (Consider how often you play/have played)", hue="controller", data=df, palette=controller_colors, order=experience_order)
plt.title("How experient do you consider your self in videogames?", pad=20)
plt.xlabel("Count")
plt.ylabel("")
plt.xlim(0, 8)  # Set x-axis limit to 0-8
plt.legend(title="Controller")
plt.tight_layout()
plt.show()
```
