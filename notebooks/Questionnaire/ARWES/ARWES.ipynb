```python
import pandas as pd

# Define the relative path to the CSV file
file_path = "../../../cleaned_data/questionnaire_df.csv"

questionnaire_df = pd.read_csv(file_path)

# Filter columns that contain "PERSONAL" or "SEQ" in their names
filtered_columns = [col for col in questionnaire_df.columns if "Controller" in col or "ARWES" in col or "participant_code" in col]

# Create a new dataframe with the filtered columns
filtered_df = questionnaire_df[filtered_columns]

# Display the new dataframe
filtered_df.head()
```

```python
import pandas as pd

# Extract relevant columns
selected_columns = filtered_df.iloc[:, :10]  # Assuming the first 9 columns are relevant

# Rename columns as per the specified format
selected_columns.columns = [
    "Controller",
    "ID",
    "SIM-WITH-VR",
    "SIM-NO-VR",
    "RW-WITH-VR-1",
    "RW-WITH-VR-2",
    "RW-WITH-VR-3",
    "RW-NO-VR-1",
    "RW-NO-VR-2",
    "RW-NO-VR-3"
]

selected_columns["Controller"] = selected_columns["Controller"].replace("Chiron", "SBC")
selected_columns.to_csv("ARWES.csv", index=False)
selected_columns
```

```python
# Function to extract the first number from a string and convert it to an integer
import re

def extract_first_number(cell):
    match = re.match(r"(\d+)", str(cell))  # Extract the first number
    return int(match.group(1)) if match else None  # Convert to int

df = selected_columns
# Apply the function to all columns except 'Controller'
columns_to_clean = [col for col in df.columns if col != "Controller" and col != "ID"]
df[columns_to_clean] = df[columns_to_clean].applymap(extract_first_number)

df.to_csv('a.csv',index=False)
```

```python
# Separate statistics for each controller (SBC and WBC)
sbc_stats = df[df["Controller"] == "SBC"].iloc[:, 1:].agg(["mean", "median", "std", "min", "max"])
wbc_stats = df[df["Controller"] == "WBC"].iloc[:, 1:].agg(["mean", "median", "std", "min", "max"])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import numpy as np

# Function to extract the first number from a string and convert it to an integer
def extract_first_number(cell):
    match = re.match(r"(\d+)", str(cell))  # Extract the first number
    return int(match.group(1)) if match else None  # Convert to int

# Apply the function to all columns except 'Controller'
columns_to_clean = [col for col in df.columns if col != "Controller"]
df[columns_to_clean] = df[columns_to_clean].applymap(extract_first_number)

# Compute the required statistics for each controller
stats_df = df.groupby("Controller")[columns_to_clean].agg(
    ["mean", "std", "median", "min", "max"]
)

# Flatten MultiIndex columns
stats_df.columns = ['_'.join(col).strip() for col in stats_df.columns.values]

# Transpose DataFrame for structured analysis
stats_df_transposed = stats_df.T

# Function to compute pooled standard deviation
def pooled_std(std_list):
    return np.sqrt(np.mean(np.array(std_list) ** 2))

# Corrected color palette size
palette_colors = ["mediumspringgreen", "mediumspringgreen", "gold", "gold"]

# Create a DataFrame for plotting
plot_df = pd.DataFrame({
    "Condition": ["With VR", "With VR", "With VR", "With VR", "Without VR", "Without VR", "Without VR", "Without VR"],
    "Controller": ["WBC SIM", "SBC SIM", "WBC RW", "SBC RW", "WBC SIM", "SBC SIM", "WBC RW", "SBC RW"],
    "Mean Score": [
        stats_df_transposed.loc["SIM-WITH-VR_mean", "WBC"],
        stats_df_transposed.loc["SIM-WITH-VR_mean", "SBC"],
        stats_df_transposed.loc[["RW-WITH-VR-1_mean", "RW-WITH-VR-2_mean", "RW-WITH-VR-3_mean"], "WBC"].mean(),
        stats_df_transposed.loc[["RW-WITH-VR-1_mean", "RW-WITH-VR-2_mean", "RW-WITH-VR-3_mean"], "SBC"].mean(),
        stats_df_transposed.loc["SIM-NO-VR_mean", "WBC"],
        stats_df_transposed.loc["SIM-NO-VR_mean", "SBC"],
        stats_df_transposed.loc[["RW-NO-VR-1_mean", "RW-NO-VR-2_mean", "RW-NO-VR-3_mean"], "WBC"].mean(),
        stats_df_transposed.loc[["RW-NO-VR-1_mean", "RW-NO-VR-2_mean", "RW-NO-VR-3_mean"], "SBC"].mean()
    ],
    "Std Dev": [
        stats_df_transposed.loc["SIM-WITH-VR_std", "WBC"],
        stats_df_transposed.loc["SIM-WITH-VR_std", "SBC"],
        pooled_std(stats_df_transposed.loc[["RW-WITH-VR-1_std", "RW-WITH-VR-2_std", "RW-WITH-VR-3_std"], "WBC"]),
        pooled_std(stats_df_transposed.loc[["RW-WITH-VR-1_std", "RW-WITH-VR-2_std", "RW-WITH-VR-3_std"], "SBC"]),
        stats_df_transposed.loc["SIM-NO-VR_std", "WBC"],
        stats_df_transposed.loc["SIM-NO-VR_std", "SBC"],
        pooled_std(stats_df_transposed.loc[["RW-NO-VR-1_std", "RW-NO-VR-2_std", "RW-NO-VR-3_std"], "WBC"]),
        pooled_std(stats_df_transposed.loc[["RW-NO-VR-1_std", "RW-NO-VR-2_std", "RW-NO-VR-3_std"], "SBC"])
    ]
})

# Create the barplot with error bars
plt.figure(figsize=(7, 4))
ax = sns.barplot(
    x="Condition", y="Mean Score", hue="Controller", data=plot_df,
    palette=palette_colors, errorbar=None
)

# Add error bars manually for each patch
for i, patch in enumerate(ax.patches):
    if i<8:
        mean = plot_df.iloc[i]["Mean Score"]
        std = plot_df.iloc[i]["Std Dev"]
        x = patch.get_x() + patch.get_width() / 2  # Center error bars correctly
        y = patch.get_height()  # Align error bars to top of the bar
        ax.errorbar(x, y, yerr=std, fmt='none', ecolor='black', capsize=5)

# Apply hatching for WBC bars
for i, patch in enumerate(ax.patches):
    if i in [0, 1, 4, 5,8,10]:
        patch.set_hatch("//")

# Customize the plot
plt.title("Mean ARWES/CSS Workload Scores in Simulation (SIM) and Real World (RW)")
plt.xlabel("Modality")
plt.ylabel("Mean Score")
plt.xticks(rotation=0)
plt.ylim(1, 7)  # Set scale from 1 to 7
plt.legend(title="Controller and Condition", loc="upper right")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import numpy as np



# Function to extract the first number from a string and convert it to an integer
def extract_first_number(cell):
    match = re.match(r"(\d+)", str(cell))  # Extract the first number
    return int(match.group(1)) if match else None  # Convert to int

# Apply the function to all columns except 'Controller'
columns_to_clean = [col for col in df.columns if col != "Controller"]
df[columns_to_clean] = df[columns_to_clean].applymap(extract_first_number)

# Compute the required statistics for each controller
stats_df = df.groupby("Controller")[columns_to_clean].agg(
    ["mean", "std", "median", "min", "max", "count"]
)

# Flatten MultiIndex columns
stats_df.columns = ['_'.join(col).strip() for col in stats_df.columns.values]

# Transpose DataFrame for structured analysis
stats_df_transposed = stats_df.T

# Function to compute standard error of the mean (SEM)
def compute_sem(std, count):
    return std / np.sqrt(count)

# Corrected color palette size
palette_colors = ["mediumspringgreen", "mediumspringgreen", "gold", "gold"]

# Create a DataFrame for plotting
plot_df = pd.DataFrame({
    "Condition": ["With VR", "With VR", "With VR", "With VR", "Without VR", "Without VR", "Without VR", "Without VR"],
    "Controller": ["WBC SIM", "SBC SIM", "WBC RW", "SBC RW", "WBC SIM", "SBC SIM", "WBC RW", "SBC RW"],
    "Mean Score": [
        stats_df_transposed.loc["SIM-WITH-VR_mean", "WBC"],
        stats_df_transposed.loc["SIM-WITH-VR_mean", "SBC"],
        stats_df_transposed.loc[["RW-WITH-VR-1_mean", "RW-WITH-VR-2_mean", "RW-WITH-VR-3_mean"], "WBC"].mean(),
        stats_df_transposed.loc[["RW-WITH-VR-1_mean", "RW-WITH-VR-2_mean", "RW-WITH-VR-3_mean"], "SBC"].mean(),
        stats_df_transposed.loc["SIM-NO-VR_mean", "WBC"],
        stats_df_transposed.loc["SIM-NO-VR_mean", "SBC"],
        stats_df_transposed.loc[["RW-NO-VR-1_mean", "RW-NO-VR-2_mean", "RW-NO-VR-3_mean"], "WBC"].mean(),
        stats_df_transposed.loc[["RW-NO-VR-1_mean", "RW-NO-VR-2_mean", "RW-NO-VR-3_mean"], "SBC"].mean()
    ],
    "SEM": [
        compute_sem(stats_df_transposed.loc["SIM-WITH-VR_std", "WBC"], stats_df_transposed.loc["SIM-WITH-VR_count", "WBC"]),
        compute_sem(stats_df_transposed.loc["SIM-WITH-VR_std", "SBC"], stats_df_transposed.loc["SIM-WITH-VR_count", "SBC"]),
        compute_sem(stats_df_transposed.loc[["RW-WITH-VR-1_std", "RW-WITH-VR-2_std", "RW-WITH-VR-3_std"], "WBC"].mean(), stats_df_transposed.loc[["RW-WITH-VR-1_count", "RW-WITH-VR-2_count", "RW-WITH-VR-3_count"], "WBC"].sum()),
        compute_sem(stats_df_transposed.loc[["RW-WITH-VR-1_std", "RW-WITH-VR-2_std", "RW-WITH-VR-3_std"], "SBC"].mean(), stats_df_transposed.loc[["RW-WITH-VR-1_count", "RW-WITH-VR-2_count", "RW-WITH-VR-3_count"], "SBC"].sum()),
        compute_sem(stats_df_transposed.loc["SIM-NO-VR_std", "WBC"], stats_df_transposed.loc["SIM-NO-VR_count", "WBC"]),
        compute_sem(stats_df_transposed.loc["SIM-NO-VR_std", "SBC"], stats_df_transposed.loc["SIM-NO-VR_count", "SBC"]),
        compute_sem(stats_df_transposed.loc[["RW-NO-VR-1_std", "RW-NO-VR-2_std", "RW-NO-VR-3_std"], "WBC"].mean(), stats_df_transposed.loc[["RW-NO-VR-1_count", "RW-NO-VR-2_count", "RW-NO-VR-3_count"], "WBC"].sum()),
        compute_sem(stats_df_transposed.loc[["RW-NO-VR-1_std", "RW-NO-VR-2_std", "RW-NO-VR-3_std"], "SBC"].mean(), stats_df_transposed.loc[["RW-NO-VR-1_count", "RW-NO-VR-2_count", "RW-NO-VR-3_count"], "SBC"].sum())
    ]
})

# Create the barplot with SEM error bars
plt.figure(figsize=(7, 4))
ax = sns.barplot(
    x="Condition", y="Mean Score", hue="Controller", data=plot_df,
    palette=palette_colors, errorbar=None
)

        
# Add error bars manually for each patch
for i, patch in enumerate(ax.patches):
    if i<8:
        mean = plot_df.iloc[i]["Mean Score"]
        sem = plot_df.iloc[i]["SEM"]
        x = patch.get_x() + patch.get_width() / 2  # Center error bars correctly
        y = patch.get_height()  # Align error bars to top of the bar
        ax.errorbar(x, y, yerr=sem, fmt='none', ecolor='black', capsize=5)

# Apply hatching for WBC bars
for i, patch in enumerate(ax.patches):
    if i in [0, 1, 4, 5,8,10]:
        patch.set_hatch("//")

# Customize the plot
plt.title("Mean ARWES/CSS Workload Scores in Simulation (SIM) and Real World (RW)")
plt.xlabel("Modality")
plt.ylabel("Mean Score")
plt.xticks(rotation=0)
plt.ylim(1, 7)  # Set scale from 1 to 7
plt.legend(title="Controller and Condition", loc="upper right")

plt.tight_layout()
output_path = "mean_arwes_sim_rw.pdf"  # Specify your desired file path
plt.savefig(output_path, format='pdf')
plt.show()
```
