```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the columns to plot
columns_to_plot = [
    "eyesight_conditions",
    "virtual_reality_experience",
    "experience_with_teleoperation",
    "driver_license",
    "handedness",
    "gender",
    "experience_with_videogames"
]

# Define colors for the controllers
colors = {"SBC": "red", "WBC": "lightgreen"}

# Set up the figure with subplots (2 per row)
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(16, 16))
axes = axes.flatten()  # Flatten to easily iterate

# Generate bar plots for each feature
for i, column in enumerate(columns_to_plot):
    ax = axes[i]

    # Get counts grouped by category and controller
    data_counts = combined_df.groupby([column, "Controller"]).size().unstack()

    # Plot bars manually to ensure color mapping is correct
    data_counts.plot(kind="bar", ax=ax, color=[colors["WBC"], colors["SBC"]])

    ax.set_title(f"Distribution of {column.replace('_', ' ').title()} by Controller")
    ax.set_xlabel(column.replace('_', ' ').title())
    ax.set_ylabel("Count")
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right")
    ax.legend(title="Controller", labels=["WBC", "SBC"])

# Remove any empty subplots if necessary
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the columns to plot
columns_to_plot = [
    "eyesight_conditions",
    "virtual_reality_experience",
    "experience_with_teleoperation",
    "driver_license",
    "handedness",
    "gender",
    "experience_with_videogames"
]

# Define colors for the controllers
colors = {"SBC": "red", "WBC": "lightgreen"}

# Define the order for experience-related categories
experience_order = ["Never", "1 or 2 times", "More than 3 times"]

# Set up the figure with subplots (2 per row)
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(10, 10))
axes = axes.flatten()  # Flatten to easily iterate

# Generate bar plots for each feature
for i, column in enumerate(columns_to_plot):
    ax = axes[i]

    # Get counts grouped by category and controller
    data_counts = combined_df.groupby([column, "Controller"]).size().unstack()

    # Ensure order is correct for experience-related columns
    if column in ["virtual_reality_experience", "experience_with_teleoperation"]:
        data_counts = data_counts.reindex(experience_order)

    # Plot bars manually to ensure color mapping is correct
    data_counts.plot(kind="bar", ax=ax, color=[colors["SBC"],colors["WBC"]] )

    ax.set_title(f"Distribution of {column.replace('_', ' ').title()} by Controller")
    ax.set_xlabel(column.replace('_', ' ').title())
    ax.set_ylabel("Count")
    ax.set_xticklabels(ax.get_xticklabels(), rotation=0, ha="right")
    ax.legend(title="Controller", labels=["SBC", "WBC"])

# Remove any empty subplots if necessary
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()

# Export the plot to a PDF
output_path = "stratification.pdf"  # Specify your desired file path
plt.savefig(output_path, format='pdf')
plt.show()
```
