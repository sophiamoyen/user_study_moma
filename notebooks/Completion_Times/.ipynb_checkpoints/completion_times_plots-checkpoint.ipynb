# 1. Transforming the data <a id='1'></a>


```python
import os
import pandas as pd

# Base path to the 'data' folder
base_path = r"E:\Hydra\sophia_ws\SerifosUpload\USER_STUDY\data"

# Initialize an empty list to store all data rows
rows = []

# Walk through the folder structure
for root, dirs, files in os.walk(base_path):
    for file in files:
        if file == "subtask_topic.csv":
            # Full path to the CSV file
            file_path = os.path.join(root, file)
            
            # Get the parent folder name that ends with "-ROSBAG"
            rosbag_folder = os.path.basename(os.path.dirname(file_path))
            
            # Determine if the file is from the "CHIRON" or "WBC" folder
            study_type = "CHIRON" if "CHIRON" in file_path else "WBC" if "WBC" in file_path else "UNKNOWN"
            
            # Determine if the folder after CHIRON/WBC is "WITH-VR" or "NO-VR"
            folder_parts = file_path.split(os.sep)
            vr_type = "UNKNOWN"
            if study_type in folder_parts:
                study_index = folder_parts.index(study_type)
                if study_index + 1 < len(folder_parts):
                    vr_type = folder_parts[study_index + 1]
            
            # Read the CSV file
            csv_data = pd.read_csv(file_path)
            
            # Extract the data column, converting it into individual columns for the DataFrame
            data_columns = csv_data['data'].tolist()  # Assuming column 'data' exists in the CSV
            
            # Create a row for this CSV
            row_dict = {'Folder': rosbag_folder, 'Controller': study_type, 'Start_Modality': vr_type}
            row_dict.update({f'Task_{i+1}': value for i, value in enumerate(data_columns)})
            
            # Append the row dictionary to the list
            rows.append(row_dict)

# Convert the list of dictionaries into a DataFrame
df = pd.DataFrame(rows)
```

So, in the middle of the study, Rickmer suggested to collect the partial time that it takes to  move the robot from the drawer to the bottle. So part of the studies have this additional subtask partial completion time and some don't. Let's organize that in the and drop it for now

```python
import numpy as np

# Function to shift values for rows where Task_6 is NaN
def shift_task_values(row):
    if np.isnan(row['Task_6']):  # Check if Task_6 is NaN
        # Shift values from Task_2 to Task_6 one column to the right
        row['Task_6'] = row['Task_5']
        row['Task_5'] = row['Task_4']
        row['Task_4'] = row['Task_3']
        row['Task_3'] = row['Task_2']
        row['Task_2'] = np.nan
    return row

# Apply the function row-wise
df = df.apply(shift_task_values, axis=1)

# Remove the column "Task_2" from the DataFrame if it exists
if 'Task_2' in df.columns:
    df = df.drop(columns=['Task_2'])
    
# Rename only the task columns to be in order from Task_1 to Task_5
new_columns = []

for col in df.columns:
    if col == 'Task_1':
        new_columns.append('Task_1')  # Keep Task_1 as is
    elif col == 'Task_3':
        new_columns.append('Task_2')
    elif col == 'Task_4':
        new_columns.append('Task_3')
    elif col == 'Task_5':
        new_columns.append('Task_4')
    elif col == 'Task_6':
        new_columns.append('Task_5')
    else:
        new_columns.append(col)  # Leave other columns unchanged

# Apply the new column names to the DataFrame
df.columns = new_columns
    
# Filter out rows where the Folder column contains "SIM"
df_cleaned = df[~df['Folder'].str.contains('SIM', na=False)]

# Export the cleaned DataFrame to the "data" folder
#df_cleaned.to_csv('../cleaned_data/completion_times_cleaned.csv', index=False)
df = df_cleaned
df
```

```python
import pandas as pd

# Creating a copy of the original DataFrame to ensure we don't overwrite the original data
df_new = df.copy()

# List of task columns (excluding Folder, Controller, Start_Modality, and Subject)
task_columns = ['Task_1', 'Task_2', 'Task_3', 'Task_4', 'Task_5']

# Start from Task_5 and go down to Task_2 for accurate subtraction
for i in range(len(task_columns)-1, 0, -1):
    df_new[task_columns[i]] = df_new[task_columns[i]] - df_new[task_columns[i - 1]]

# Task_1 should stay as it is (no subtraction needed)
df_new['Task_1'] = df_new['Task_1']

# Show the resulting DataFrame with the task times for each task
df = df_new

# Drop rows where any column contains "SIM"
df = df[~df.apply(lambda row: row.astype(str).str.contains('SIM')).any(axis=1)]

# Reindex the DataFrame after dropping rows
df = df.reset_index(drop=True)

df
```

# 2. Analysing data for one subject <a id='2'></a>

```python
import pandas as pd
import matplotlib.pyplot as plt
import random

# Define the task names
task_names = {
    "Task_1": "Open Drawer",
    "Task_2": "Go to and pick up bottle",
    "Task_3": "Drop the bottle inside the drawer",
    "Task_4": "Close the drawer",
    "Task_5": "Park the robot in the parking space"
}

# Extract subject code, trial, and modality information
df['Subject'] = df['Folder'].str.extract(r'([A-Z]{4}\d{2})')
df['Trial'] = df['Folder'].str.extract(r'TRIAL-(\d)').astype(int)
df['Modality'] = df['Folder'].str.extract(r'(WITH-VR|NO-VR)')

# Function to plot for a specified subject
def plot_subject_data(subject_code):
    # Filter data for the selected subject
    subject_data = df[df['Subject'] == subject_code]
    if subject_data.empty:
        print(f"No data found for subject: {subject_code}")
        return

    tasks = [f"Task_{i}" for i in range(1, 6)]

    # Plot each task for the specified subject
    for task in tasks:
        plt.figure(figsize=(8, 5))

        # Separate data for WITH-VR and NO-VR
        with_vr_data = subject_data[subject_data['Modality'] == 'WITH-VR'].sort_values('Trial')
        no_vr_data = subject_data[subject_data['Modality'] == 'NO-VR'].sort_values('Trial')

        # Ensure data are 1D arrays
        with_vr_trials = with_vr_data['Trial'].values
        with_vr_times = with_vr_data[task].values

        no_vr_trials = no_vr_data['Trial'].values
        no_vr_times = no_vr_data[task].values

        # Get task-specific maximum to set the same scale for both modalities
        task_max = max(with_vr_times.max() if len(with_vr_times) > 0 else 0,
                       no_vr_times.max() if len(no_vr_times) > 0 else 0)

        # Plotting data for both modalities
        plt.plot(with_vr_trials, with_vr_times, marker='o', color='blue', label='With VR')
        plt.plot(no_vr_trials, no_vr_times, marker='o', color='red', label='Without VR')

        # Customizing the plot
        plt.xlabel("Trial")
        plt.ylabel("Completion Time (s)")
        plt.ylim(0, task_max + 10)  # Same scale for both modalities with buffer
        
        # Set the x-axis ticks to show only the integers for trials
        plt.xticks([1, 2, 3])  # Set trial numbers explicitly
        
        # Set the title using the mapped task names
        plt.title(f"{subject_code}: {task_names[task]}")

        plt.legend()
        plt.grid(True)

        # Show the plot
        plt.tight_layout()
        plt.show()

# List of all subject codes
subjects = df['Subject'].unique()

# Randomly select one subject
random_subject = random.choice(subjects)

# Example usage: Plot data for the randomly selected subject
plot_subject_data(random_subject)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import random

# Extract subject code, trial, and modality information
df['Subject'] = df['Folder'].str.extract(r'([A-Z]{4}\d{2})')
df['Trial'] = df['Folder'].str.extract(r'TRIAL-(\d)').astype(int)
df['Modality'] = df['Folder'].str.extract(r'(WITH-VR|NO-VR)')

# Sum the task completion times to create a "Total Time" column (sum of Task 1 to Task 5)
df['Total Time'] = df[['Task_1', 'Task_2', 'Task_3', 'Task_4', 'Task_5']].sum(axis=1)

# List of all subject codes
subjects = df['Subject'].unique()

# Randomly select one subject
random_subject = random.choice(subjects)

# Function to plot total time for the selected subject
def plot_total_time(subject_code):
    # Filter data for the selected subject
    subject_data = df[df['Subject'] == subject_code]
    if subject_data.empty:
        print(f"No data found for subject: {subject_code}")
        return

    # Separate data for WITH-VR and NO-VR
    with_vr_data = subject_data[subject_data['Modality'] == 'WITH-VR'].sort_values('Trial')
    no_vr_data = subject_data[subject_data['Modality'] == 'NO-VR'].sort_values('Trial')

    # Ensure we have trial numbers as the x-axis (1, 2, 3)
    trial_numbers_with_vr = with_vr_data['Trial'].values
    total_time_with_vr = with_vr_data['Total Time'].values

    trial_numbers_no_vr = no_vr_data['Trial'].values
    total_time_no_vr = no_vr_data['Total Time'].values

    # Plotting data for both modalities
    plt.figure(figsize=(8, 5))

    plt.plot(trial_numbers_with_vr, total_time_with_vr, marker='o', color='blue', label='With VR')
    plt.plot(trial_numbers_no_vr, total_time_no_vr, marker='o', color='red', label='Without VR')

    # Customizing the plot
    plt.xlabel("Trial")
    plt.ylabel("Total Completion Time (s)")
    plt.xticks([1, 2, 3])  # Set trial numbers explicitly
    plt.title(f"{subject_code}: Total Completion Time Across Trials")
    plt.legend()
    plt.grid(True)

    # Show the plot
    plt.tight_layout()
    plt.show()

# Example usage: Plot data for the randomly selected subject
plot_total_time(random_subject)
```
